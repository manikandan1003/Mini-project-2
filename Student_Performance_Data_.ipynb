```python
# Replace NaN in 'company_name' with 'Not Placed' where 'placement_status' is 'Not Ready'
df.loc[(df['placement_status'] == 'Not Ready') & (df['company_name'].isna()), 'company_name'] = 'Not Placed'
df.loc[(df['placement_status'] == 'Ready') & (df['company_name'].isna()), 'company_name'] = 'Document verification'
```

```python
# SQL query to count total students
sql = "SELECT COUNT(*) FROM student_performance"

# Execute query
mycursor.execute(sql)

# Fetch result
total_students = mycursor.fetchone()[0]
print(f"Total number of students: {total_students}")
```

```python
# Step 2: SQL query to group by placement_status
query = """
    SELECT placement_status, COUNT(*) AS count
    FROM student_performance
    GROUP BY placement_status
"""

# Step 3: Execute and fetch results
mycursor.execute(query)
results = mycursor.fetchall()

# Step 4: Print the result
print("Placement Summary:")
for status, count in results:
    print(f"{status}: {count} student(s)")
```

```python
# Step 3: SQL to find city with highest placements
query = """
    SELECT city, COUNT(*) AS placed_count
    FROM student_performance
    WHERE placement_status = 'Placed'
    GROUP BY city
    ORDER BY placed_count DESC
    LIMIT 1
"""

mycursor.execute(query)
result = mycursor.fetchall()

# Step 3: Display result as table
print(tabulate(result, headers=["City", "Placed Students"], tablefmt="grid"))
```

```python
# Step 2: SQL to calculate average placement package
query = """
    SELECT AVG(placement_package) AS average_package
    FROM student_performance
    WHERE placement_status = 'Placed'
"""

mycursor.execute(query)
result = mycursor.fetchone()

# Step 3: Print the result
average = result[0]
if average is not None:
    print(f"Average Placement Package: ₹{average:.2f}")
else:
    print("No placed students found.")
```

```python
# Step 2: SQL Query to count students per graduation year
query = """
    SELECT graduation_year, COUNT(*) AS total_students
    FROM student_performance
    GROUP BY graduation_year
    ORDER BY graduation_year
"""

mycursor.execute(query)
results = mycursor.fetchall()

# Step 3: Display result as table
print(tabulate(results, headers=["Graduation Year", "Number of Students"], tablefmt="grid"))
```

```python
# Step 2: SQL Query to count placed students by gender
query = """
    SELECT gender, COUNT(*) AS placed_count
    FROM student_performance
    WHERE placement_status = 'Placed'
    GROUP BY gender
"""

mycursor.execute(query)
results = mycursor.fetchall()

# Step 3: Display result as table
print(tabulate(results, headers=["Gender", "Number of Placed Students"], tablefmt="grid"))
```

```python
query = """
    SELECT internships_completed,
           SUM(CASE WHEN placement_status = 'Placed' THEN 1 ELSE 0 END) AS placed_students,
           SUM(CASE WHEN placement_status != 'Placed' THEN 1 ELSE 0 END) AS not_placed_students
    FROM student_performance
    GROUP BY internships_completed
    ORDER BY internships_completed
"""

mycursor.execute(query)
results = mycursor.fetchall()

# Step 3: Display result in table format
print(tabulate(results, headers=["Internships Completed", "Placed Students", "Not Placed Students"], tablefmt="grid"))
```

```python
# Step 2: SQL query to calculate average certifications of placed students
query = """
    SELECT AVG(certifications_earned)
    FROM student_performance
    WHERE placement_status = 'Placed'
"""

mycursor.execute(query)
result = mycursor.fetchone()

# Step 3: Format and display result as a table
average_certifications = round(result[0], 2) if result[0] is not None else 0
table = [["Placed Students", average_certifications]]

print(tabulate(table, headers=["Status", "Average Certifications Earned"], tablefmt="grid"))
```

```python
# Step 2: SQL Query to get average latest_project_score per course_batch
query = """
    SELECT course_batch, ROUND(AVG(latest_project_score), 2) AS avg_score
    FROM student_performance
    GROUP BY course_batch
    ORDER BY avg_score DESC
"""

mycursor.execute(query)
results = mycursor.fetchall()

# Step 3: Display in table format
print(tabulate(results, headers=["Course Batch", "Avg. Latest Project Score"], tablefmt="grid"))
```

```python
query = """
    SELECT AVG(mock_interview_score)
    FROM student_performance
    WHERE placement_status = 'Placed'
"""

mycursor.execute(query)
result = mycursor.fetchone()

# Step 3: Format result
average_mock_score = round(result[0], 2) if result[0] is not None else 0
table = [["Placed Students", average_mock_score]]

# Step 4: Display in table format
print(tabulate(table, headers=["Status", "Average Mock Interview Score"], tablefmt="grid"))
```

```python
# Step 2: SQL query to count not placed students by gender
query = """
    SELECT gender, COUNT(*) AS not_placed_count
    FROM student_performance
    WHERE placement_status != 'Placed'
    GROUP BY gender
"""

mycursor.execute(query)
results = mycursor.fetchall()

# Step 3: Display result in table format
print(tabulate(results, headers=["Gender", "Not Placed Students"], tablefmt="grid"))
```
